```python
# Dtype (data types)
x = np.array([1, 2, 3], dtype = np.float32)
print(x.dtype)

y = x.astype(np.int64)
print(y.dtype)
```

```python
# Inspecting arrays

m = np.arange(12).reshape(3,4)
print(m)

print(m.ndim) # 2D arrayb
print(m.shape)
print(m.size)
print(m.dtype)
```

```python
# Indexing and Slicing
A = np.array([[10, 11, 12],
             [20, 21, 22],
             [30, 31, 32]])
print(A[0, 1]) # 11
print(A[1])
print(A[:, 0]) 
print(A[0:2, 1:3])
```

```python
# Fancy Indexing (by Integer arrays)
w = np.array([10, 20, 30, 40, 50])
idx = np.array([4, 0, 3])
print(w[idx])
```

```python
# Vectorized Operations (element-wise)
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print(a + b)
print(a * b)
print(b / a)
print(a ** 2)
```

```python
# Broadcasting
M = np.arange(6).reshape(2,3)
print(M)
V = np.array([10, 100, 1000])


print(M + V)

col = np.array([[1], [2]])
print(col)

print( M + col)
```

```python
# Universal Functions (ufuncs)

x = np.array([1, 4, 9, 16], dtype = float)
print(np.sqrt(x))
print(np.log(x))
print(np.maximum(x, 5))

#Try
x = [-1, 0, 1]
print(np.exp(x))
print(np.exp(x) / (1 + np.exp(x)))
```
