```python
#setup
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Two ways to use matplotlib 

plt.plot([1,2,3], [2, 3, 4]) #line
plt.title("Quick Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
#Object-oriented (clean & scalable)
fig, ax = plt.subplots() #fig: whole canvas, ax: one plot area
ax.plot([1,2, 3], [2, 4, 3])
ax.set_title("OO Line")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True)
plt.show()
```

```python
# Line plots - trends over an ordered x- axis

x = [1, 2, 3, 4]
y = [2, 3, 5, 8]

plt.plot(x, y, marker = 'o', linestyle = '--', linewidth = 2)
plt.title("Line with Markers")
plt.xlabel("Step")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
# Scatter plots - relations between two numbers
x = [1, 2, 3, 4, 5]
y = [1.2, 1.9, 3.1, 3.9, 5.2]

plt.scatter(x, y, s = 60, alpha = 0.9)
plt.title("Scatter")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# Bar charts = compare categories
items = ["A", "B", "C"]
counts = [5, 8, 3]

plt.bar(items, counts)
plt.title("Bar Chart")
plt.xlabel("Item")
plt.ylabel("Count")
plt.grid(True)
plt.show()

# Horizontal bars
plt.barh(items, counts)
plt.title("Horizontal Bars")
plt.xlabel("Counts")
plt.ylabel("Item")
plt.show()
```
