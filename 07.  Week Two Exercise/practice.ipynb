```python
# Problem 6: Function in Python
def calculate_bmi(weight, height):
    bmi = weight / (height**2)
    return bmi

def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

weight = 70
height = 1.75
print(f"BMI: {calculate_bmi(weight,height)}, Category: {categorize_bmi(calculate_bmi(weight,height))}")
```

```python
# Probelm 7: Testing and Error Handing in Python

def safe_divide(numerator, denominator):
    try:
        result = numerator /denominator
        
    except ZeroDivisionError:
        print('Error: Division by zero is no allowed.')
        return None

    else:
        return result



print('Result 1:', safe_divide(10,2))
# print('Result 1:', safe_divide(10,0))
    
```

```python
# Problem 8: Testing and Handing Errors
def divide_number(a,b):
    try:
        if not (isinstance(a,(int,float)) and isinstance(b, (int,float))):
            raise TypeError("Error: Input must be numbers")
        if b == 0:
            raise ZeroDivisionError("Error: Division by zero is not allowed")
        return a/b
    except (TypeError, ZeroDivisionError) as e:
        return e

print(divide_number(10,2))   
print(divide_number('a',2))   
print(divide_number(10,0))   

        
```
